```python
pd.get_dummies(Tdata["Sex"])
```

```python
sex=pd.get_dummies(Tdata["Sex"],drop_first=True)
sex.head()
```

```python
embark=pd.get_dummies(Tdata["Embarked"],drop_first=True)
embark.head()
```

```python
pcl=pd.get_dummies(Tdata["Pclass"],drop_first=True)
pcl.head()
```

```python
Tdata=pd.concat([Tdata,sex,embark,pcl],axis=1)
```

```python
Tdata.head()
```

```python
Tdata.head()
```

```python
Tdata.drop("PassengerID",axis=1,inplace=True)
```

```python
Tdata.head()
```

```python
X=Tdata.drop("Survived",axis=1)
y=Tdata["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
classification_report(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```
